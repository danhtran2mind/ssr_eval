```python
import pandas as pd 
import os
import numpy as np 
import json  

def load_json(fname):
    with open(fname,'r') as f:
        data = json.load(f)
        return data

FINAL_RESULTS="/vol/research/dcase2022/sr_eval_vctk/final_result"
METRICS="lsd"
```

```python
def rename_key(dic):
    ret = {}
    for k in dic.keys():
        index = k.rfind("_")
        ret[k[:index]] = dic[k] 
    return ret
```

```python
final_result = {}
settings = None
for result in os.listdir(FINAL_RESULTS):
    test_name = result.split("-")[-1].split(".")[0]
    averaged_result = rename_key(load_json(os.path.join(FINAL_RESULTS, result))['averaged'])
    for k in averaged_result.keys():
        if(settings is None): settings = list(averaged_result.keys())
        if(len(averaged_result.keys()) > len(settings)): settings = list(averaged_result.keys())
        averaged_result[k] =averaged_result[k][METRICS]
    final_result[test_name] = averaged_result
```

```python
final_result['WSRGlow_2_ratio'].keys()
```

```python
result_list = {}
trail_list = []
for trail in final_result.keys():
    trail_list.append(trail)
    for k in settings:
        if(k not in result_list.keys()): result_list[k] = []
        if(k not in final_result[trail].keys()): 
            # if("Glow" in trail): print(trail, k)
            result_list[k].append(0)
        else: result_list[k].append(final_result[trail][k])
    
```

```python
df = pd.DataFrame.from_dict(result_list)
index = {}
for i in range(len(trail_list)):
    index[i] = trail_list[i]
df_new = df.rename(index=index)
```
